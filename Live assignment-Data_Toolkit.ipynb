3. Explain the following terms:
a) The difference in np.array, np.asarray and np.asanyarray
b) The difference between Deep copy and shallow copy

```python
"""
=> np.array:

* np.array always creates a new copy of the input data, regardless of whether the input
  is already a NumPy array or a different type of sequence (like a list or a tuple).
* It is most commonly used when you want to ensure that you have a NumPy array,
  even if the input is already an array.
  
=> np.asarray:

* np.asarray creates a new array if the input is not already an array.
  If the input is already a NumPy array, it returns a view of the same data
  by default (unless the copy parameter is explicitly set to True).
* This function is useful when you want to avoid unnecessary copying of data.
  If the input is already a NumPy array, np.asarray is preferred over np.array
  because it can save memory by returning a view instead of a copy.
  
=> np.asanyarray:

* np.asanyarray also creates an array, but it tries to preserve the input type if possible.
  If the input is already a NumPy array, it returns a view of the same data.
  However, if the input is a subclass of np.ndarray, it preserves that subclass type.
* This function is useful when you want to ensure that you have an array,
  but you also want to preserve the type of the input if it's a subclass of np.ndarray.
  
"""
import numpy as np

# Original data (Python list)
data = [1, 2, 3, 4, 5]

# np.array always creates a new array
arr1 = np.array(data)
print(arr1 is data)  # False (new array created)

# np.asarray creates a view if input is already a NumPy array
arr2 = np.asarray(arr1)
print(arr2 is arr1)  # True (view of the same data)

# np.asanyarray preserves input type if it's a subclass of np.ndarray
sub_class_array = np.ma.masked_array([1, 2, 3, 4, 5])
arr3 = np.asanyarray(sub_class_array)
print(isinstance(arr3, np.ma.masked_array))  # True (preserves subclass type)
```

```python
"""
=> Shallow Copy:

* In a shallow copy, a new object is created, but instead of copying the elements themselves,
  references to the original elements are copied. This means that the new object refers to
  the same memory locations as the original object for its data.
* Modifications made to the elements of the new object will affect the
  corresponding elements in the original object, and vice versa.
* In NumPy, the view method creates a shallow copy. Slicing an array also creates a shallow copy.
* Shallow copies are generally faster and use less memory because they don't duplicate the data.

=> Deep Copy:

* In a deep copy, a new object is created, and a completely new set of elements is created
  and populated with copies of the original elements.
  This means that the new object has its own independent copy of the data.
* Modifications made to the elements of the new object will not affect the
  corresponding elements in the original object, and vice versa.
* In NumPy, the copy method creates a deep copy.
* Deep copies are generally slower and use more memory because they duplicate the data.

"""

import numpy as np

# Original array
original_array = np.array([[1, 2], [3, 4]])

print("Original array:")
print(original_array)

# Shallow copy using view
shallow_copy = original_array.view()

# Deep copy using copy
deep_copy = original_array.copy()

# Modify the original array
original_array[0, 0] = 99

# Changes in the shallow copy affect the original array
print("\nShallow copy:")
print(shallow_copy)

# Changes in the deep copy do not affect the original array
print("\nDeep copy:")
print(deep_copy)
```

7. Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type. Also display the table attributes and data types of each column.

```python
import pandas as pd 
import numpy as np
df = pd.read_csv('People Data.csv')

def clean_phone(phone):
    if pd.isna(phone) or phone == '':
        return np.nan
    # Remove non-numeric characters
    cleaned_phone = ''.join(filter(str.isdigit, phone))
    # Return NaN if cleaned_phone is empty, otherwise return the cleaned phone number
    return np.nan if cleaned_phone == '' else int(cleaned_phone)

# Apply the cleaning function to the 'Phone' column
df['Phone'] = df['Phone'].apply(clean_phone)

# Convert the 'Phone' column to Int64 to handle integers with NaNs
df['Phone'] = df['Phone'].astype('Int64')

# Configure pandas to display numbers in full without scientific notation
pd.set_option('display.float_format', '{:.0f}'.format)

# Display the DataFrame and its data types
print("DATAFRAME : \n",df)
print("\nData Types :\n",df.dtypes)
```

15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
arr = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

#  Plot the histogram
plt.hist(arr, bins=30, density=True, alpha=0.6, color='g')

# Overlay the PDF
# Calculate the mean and standard deviation of the data
mu, std = np.mean(arr), np.std(arr)
# Generate x values
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
# Calculate the PDF
p = norm.pdf(x, mu, std)
# Plot the PDF
plt.plot(x, p, 'k', linewidth=2)

# Label the axes and set the title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

plt.show()
```

17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Q1'
    elif x < 0 and y > 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    else:
        return 'Q4'

quadrants = [get_quadrant(xi, yi) for xi, yi in zip(x, y)]

df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Plot the scatter plot using Seaborn
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='deep')

#  Add legend, label the axes, and set the title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Quadrant-wise Scatter Plot')
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)
plt.legend(title='Quadrant')

plt.show()
```

18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Ensure Bokeh plots display in the notebook
output_notebook()

# Generate the sine wave data
x = np.linspace(0, 4 * np.pi, 1000)
y = np.sin(x)

# Create a Bokeh plot
# Initialize a Bokeh figure
p = figure(title="Sine Wave Function", width=800, height=400)

# Plot the sine wave line
p.line(x, y, line_width=2, color='blue')

# Add grid lines, labels, and title
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'
p.xaxis.axis_label = 'X Axis'
p.yaxis.axis_label = 'Y Axis'

show(p)

```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256

# Ensure Bokeh plots display in the notebook
output_notebook()

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Initialize a Bokeh figure
p = figure(x_range=categories, title="Random Categorical Bar Chart", 
           toolbar_location=None, tools="", width=800, height=400)

# Plot the bar chart
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=factor_cmap('categories', palette=Viridis256, factors=categories))

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Category'
p.yaxis.axis_label = 'Value'

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

```python
import numpy as np
import plotly.graph_objs as go
import plotly.io as pio

np.random.seed(0)  # For reproducibility
x = np.arange(0, 100)
y = np.random.randn(100).cumsum()  # Cumulative sum to simulate a trend

# Create the Plotly plot
fig = go.Figure()

# Add the line plot
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Label the axes and set the title
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis',
    yaxis_title='Y Axis'
)

pio.show(fig)
```
